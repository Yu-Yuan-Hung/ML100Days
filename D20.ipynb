```python
import numpy as np
import seaborn as sns

#設定圖形樣式 - whitegrid
sns.set_style('whitegrid')

# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
x = np.random.normal(2, 1, 75)
y = 2 + 1.5 * x + np.random.normal(0, 2, 75)


# 畫圖
sns.residplot(x,y)
```

```python
# bin: 指的是特徵值, 
# kde: on/off
sns.distplot(x, bins=20, kde=True, rug=True);
```
